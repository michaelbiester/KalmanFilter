## Definitions

**definition / multivariate normal distribution**



$$
p(\mathbf{x}; \mathbf{\mu}, \mathbf{\Sigma}) = \frac{1}{\sqrt{(2 \pi)^n |\mathbf{\Sigma}|}} \cdot exp \left( -\frac{1}{2} \left(\mathbf{x} - \mathbf{\mu}\right)^T \cdot \mathbf{\Sigma}^{-1} \cdot \left(\mathbf{x} - \mathbf{\mu}\right) \right)
$$

| property / symbol   | description | dimension |
|---------------------|-------------|-----------|
| $p(\mathbf{x}; \mathbf{\mu}, \mathbf{\Sigma})$ | probability density function | real valued function |
| $\mathbf{x}$        | multivariate random vector | $ \in \mathbb{R}^{n \times 1}$ |
| $\mathbf{\mu}$      | vector of mean values | $ \in \mathbb{R}^{n \times 1}$ |
| $\mathbf{\Sigma}$   | covariance matrix     | $ \in \mathbb{R}^{n \times n}$ |
| $|\mathbf{\Sigma}|$ | determinant of covariance matrix | scalar |
| $\mathbf{\Sigma}^{-1}$ | inverse of covariance matrix | $ \in \mathbb{R}^{n \times n}$ |


**definition / univariate normal distribution**

$$
p(x; \mu, \sigma) =  \frac{1}{\sqrt{2 \pi} \sigma} \cdot exp \left( -\frac{1}{2 \sigma^2} \left(x - \mu\right)^2 \right)
$$



| property / symbol   | description | 
|---------------------|-------------|
| $p(x; \mu, \sigma)$ | probability density function |
| $x$        | random variable | 
| $\mu$      | mean value of $x$ | 
| $\sigma$   | standard deviation of $x$    | 


## Properties of the covariance matrix

The covariance matrix $\mathbf{\Sigma}$ is symmetric and positive definite. 

**proof / symmetry**

For $n$ random variables $X_1,\ \ldots,\ X:n$ the covariance matrix is defined by:

$$
\Sigma =\left[\begin{array}{ccc}
Cov(X_1, X_1) & \cdots & Cov(X_1, X_n) \\
\vdots & \ddots & \vdots \\
Cov(X_n, X_1) & \cdots & Cov(X_n, X_n)
\end{array}\right]
$$

For the individual covariance of random variables $X$ and $Y$ the definition is:

$$\begin{align}
Cov(X, Y) &= E\left( \left(X - E(X) \right) \cdot \left(Y - E(Y) \right) \right) \\
&= E\left( \left(X - \mu_x \right) \cdot \left(Y - \mu_y \right) \right) \\
&= E\left( X\cdot Y \right) - 2 \cdot \mu_x \cdot \mu_y + \mu_x \cdot \mu_y  \\
&= E\left( X\cdot Y \right) - \mu_x \cdot \mu_y 
\end{align}
$$

$$
\Sigma =\left[\begin{array}{ccc}
E\left( \left(X_1 - E(X_1) \right) \cdot \left(X_1 - E(X_1) \right) \right) & \cdots & E\left( \left(X_1 - E(X_1) \right) \cdot \left(X_n - E(X_n) \right) \right) \\
\vdots & \ddots & \vdots \\
E\left( \left(X_n - E(X_n) \right) \cdot \left(X_1 - E(X_1) \right) \right) & \cdots & E\left( \left(X_n - E(X_n) \right) \cdot \left(X_n - E(X_n) \right) \right)
\end{array}\right]
$$

Since expectation is a linear operation we get:

$$\begin{align}
\Sigma &= E\left(\left[\begin{array}{ccc}
\left(X_1 - E(X_1) \right) \cdot \left(X_1 - E(X_1) \right) & \cdots & \left(X_1 - E(X_1) \right) \cdot \left(X_n - E(X_n) \right) \\
\vdots & \ddots & \vdots \\
\left(X_n - E(X_n) \right) \cdot \left(X_1 - E(X_1) \right) & \cdots & \left(X_n - E(X_n) \right) \cdot \left(X_n - E(X_n) \right) 
\end{array}\right] \right) \\
&= E\left( \left[\begin{array}{c}
\left(X_1 - E(X_1) \right) \\
\vdots \\
\left(X_n - E(X_n) \right)
\end{array}\right] \cdot \left[\begin{array}{ccc}
\left(X_1 - E(X_1) \right) & \cdots & \left(X_n - E(X_n) \right)
\end{array}\right] 
\right) \\
&= E\left( \left(\mathbf{x} - \mathbf{\mu} \right) \cdot \left(\mathbf{x} - \mathbf{\mu} \right)^T  \right)
\end{align}
$$

Since $\left(\mathbf{x} - \mathbf{\mu} \right) \cdot \left(\mathbf{x} - \mathbf{\mu} \right)^T$ is symmetric the expectation is symmetric too. And so is the covariance matrix $\Sigma$.

**proof / positive definiteness**

For the covariance matrix defined by equation 

$$
\Sigma = E\left( \left(\mathbf{x} - \mathbf{\mu} \right) \cdot \left(\mathbf{x} - \mathbf{\mu} \right)^T  \right)
$$

we need to show

$$
\mathbf{z}^T \cdot \Sigma \cdot \mathbf{z} \gt 0; \ for \ \mathbf{z} \neq \mathbf{0}
$$

for notational convenience we define the vector $\mathbf{y}$:

$$
\mathbf{y} = \mathbf{x} - \mathbf{\mu}
$$

Then 

$$\begin{align} 
\mathbf{z}^T \cdot \Sigma \cdot \mathbf{z} &= \mathbf{z}^T \cdot E\left( \mathbf{y} \cdot \mathbf{y}^T \right) \cdot \mathbf{z} \\
&= E\left( \mathbf{z}^T \cdot \mathbf{y} \cdot \mathbf{y}^T \cdot \mathbf{z} \right) \\
&= E\left( \left(\mathbf{z}^T \cdot \mathbf{y}\right) \cdot \left(\mathbf{y}^T \cdot \mathbf{z}\right) \right) \\
&= E \left( \left(\mathbf{z}^T \cdot \mathbf{y}\right)^2 \right)
\end{align}
$$

Expression $\left(\mathbf{z}^T \cdot \mathbf{y}\right)^2$ is scalar and $\ge 0$. The expectation is $\ge 0$ too. So we have shown that the covariance matrix is at least positive semidefinite.

## Normal distribution in 2 Dimensions

$$
p(\mathbf{x}; \mathbf{\mu}, \mathbf{\Sigma}) = \frac{1}{\sqrt{(2 \pi)^2 |\mathbf{\Sigma}|}} \cdot exp \left( -\frac{1}{2} \left(\mathbf{x} - \mathbf{\mu}\right)^T \cdot \mathbf{\Sigma}^{-1} \cdot \left(\mathbf{x} - \mathbf{\mu}\right) \right)
$$

The simplest case is analysed first.

We assume a covariance matrix $\mathbf{\Sigma}$ which has only entries in the main diagonal.

$$
\mathbf{\Sigma} = \left[\begin{array}{cc}
\sigma_1^2 & 0 \\
0 & \sigma_2^2
\end{array}\right]
$$

The inverse covariance matrix $\mathbf{\Sigma}^{-1}$ is then a diagonal matrix too:

$$
\mathbf{\Sigma}^{-1} = \left[\begin{array}{cc}
\frac{1}{\sigma_1^2} & 0 \\
0 & \frac{1}{\sigma_2^2}
\end{array}\right]
$$

For the normalising factor $\frac{1}{\sqrt{(2 \pi)^2 |\mathbf{\Sigma}|}}$ we obtain:

$$
\frac{1}{\sqrt{(2 \pi)^2 |\mathbf{\Sigma}|}} = \frac{1}{2 \pi \cdot \sigma_1 \cdot \sigma_2 }
$$

With 

$$
\left(\mathbf{x} - \mathbf{\mu}\right) = \left[\begin{array}{cc}
x_1 - \mu_1 \\ x_2 - \mu_2
\end{array}\right]
$$

the expression of the exponent is computed as:

$$\begin{align}
-\frac{1}{2} \left(\mathbf{x} - \mathbf{\mu}\right)^T \cdot \mathbf{\Sigma}^{-1} \cdot \left(\mathbf{x} - \mathbf{\mu}\right) &= -\frac{1}{2} \left[\begin{array}{cc}
x_1 - \mu_1 & x_2 - \mu_2
\end{array}\right] \cdot \left[\begin{array}{cc}
\frac{1}{\sigma_1^2} & 0 \\
0 & \frac{1}{\sigma_2^2}
\end{array}\right] \cdot \left[\begin{array}{c}
x_1 - \mu_1 \\ x_2 - \mu_2
\end{array}\right] \\
&= -\frac{1}{2} \left[\begin{array}{cc}
x_1 - \mu_1 & x_2 - \mu_2
\end{array}\right] \cdot  \left[\begin{array}{c}
\frac{1}{\sigma_1^2} \cdot (x_1 - \mu_1) \\ \frac{1}{\sigma_2^2} \cdot (x_2 - \mu_2)
\end{array}\right] \\
&= -\frac{1}{2} \left(\frac{(x_1 - \mu_1)^2}{\sigma_1^2}  + \frac{(x_2 - \mu_2)^2}{\sigma_2^2} \right)
\end{align}
$$

Putting everthing together we get an expression for the *bivariate* normal distribution $p(x_1, x_2; \mu_1, \mu_2, \sigma_1, \sigma_2)$

$$\begin{align}
p(x_1, x_2; \mu_1, \mu_2, \sigma_1, \sigma_2) &= \frac{1}{2 \pi \cdot \sigma_1 \cdot \sigma_2 } \cdot exp \left( -\frac{1}{2} \left(\frac{(x_1 - \mu_1)^2}{\sigma_1^2}  + \frac{(x_2 - \mu_2)^2}{\sigma_2^2} \right) \right) \\
&= \frac{1}{\sqrt{2 \pi } \cdot \sigma_1 } \cdot exp \left( -\frac{1}{2 \sigma_1^2} (x_1 - \mu_1)^2 \right) \cdot \frac{1}{\sqrt{2 \pi } \cdot \sigma_2  } \cdot exp \left( -\frac{1}{2 \sigma_2^2} (x_2 - \mu_2)^2 \right)
\end{align}
$$

Since the covariance is a diagonal matrix the 2D normal distribution is just the product of two 1D normal distributions.

## Iso Contours / Bivariate Normal Distribution

Again assuming a diagonal covariance matrix there are applications where we need to know the set of $x_1,x_2$ values for which the probability density function has a constant value $c$.

$$
c = p(x_1, x_2; \mu_1, \mu_2, \sigma_1, \sigma_2)
$$

$$\begin{align}
c &= \frac{1}{\sqrt{2 \pi } \cdot \sigma_1 } \cdot exp \left( -\frac{1}{2 \sigma_1^2} (x_1 - \mu_1)^2 \right) \cdot \frac{1}{\sqrt{2 \pi } \cdot \sigma_2  } \cdot exp \left( -\frac{1}{2 \sigma_2^2} (x_2 - \mu_2)^2 \right) \\
c \cdot 2 \pi \cdot \sigma_1 \cdot \sigma_2 &= exp \left( -\frac{1}{2 \sigma_1^2} (x_1 - \mu_1)^2 \right) \cdot  exp \left( -\frac{1}{2 \sigma_2^2} (x_2 - \mu_2)^2 \right) \\
\ln{(c \cdot 2 \pi \cdot \sigma_1 \cdot \sigma_2)} &= -\frac{1}{2 \sigma_1^2} (x_1 - \mu_1)^2   -\frac{1}{2 \sigma_2^2} (x_2 - \mu_2)^2 \\
\ln{(\frac{1}{c \cdot 2 \pi \cdot \sigma_1 \cdot \sigma_2})} &= \frac{(x_1 - \mu_1)^2 }{2 \sigma_1^2} +  \frac{(x_2 - \mu_2)^2 }{2 \sigma_2^2} \\
1 &= \frac{(x_1 - \mu_1)^2 }{2 \sigma_1^2 \cdot \ln{(\frac{1}{c \cdot 2 \pi \cdot \sigma_1 \cdot \sigma_2})}} +  \frac{(x_2 - \mu_2)^2 }{2 \sigma_2^2 \cdot \ln{(\frac{1}{c \cdot 2 \pi \cdot \sigma_1 \cdot \sigma_2})}}
\end{align}
$$

Introducing variables $r_1$ and $r_2$ 

$$\begin{align}
r_1^2 &= 2 \sigma_1^2 \cdot \ln{(\frac{1}{c \cdot 2 \pi \cdot \sigma_1 \cdot \sigma_2})} \\
r_2^2 &= 2 \sigma_2^2 \cdot \ln{(\frac{1}{c \cdot 2 \pi \cdot \sigma_1 \cdot \sigma_2})}
\end{align}
$$

yields the equation of an ellipse centered at $x_1=\mu_1;\ x_2=\mu_2$ :


$$
1 = \frac{(x_1 - \mu_1)^2 }{r_1^2} +  \frac{(x_2 - \mu_2)^2 }{r_2^2}
$$

$r_1$ and $r_2$ are the half axes in $x_1$ and $x_2$ direction as can be shown if we use the equation of an ellipse centered at $0,0$.
The equation of the ellipse simplifies to:


$$
1 = \frac{x_1^2}{r_1^2} +  \frac{x_2^2 }{r_2^2}
$$

Setting $x_2=0$ we get $\pm r_1 = x_1$. 

And setting $x_1=0$ we get $\pm r_2 = x_2$

From the definitions of $r_1$ and $r_2$ it follows that half axes $r_1$ and $r_2$ are proportional to $\sigma_1$ and $\sigma_2$ respectively.


$$
\frac{r_1}{r_2} = \frac{\sigma_1}{\sigma_2}
$$

```python
import numpy as np
import scipy.stats as stats
import scipy.linalg as linalg
import matplotlib.pyplot as plt
```

```python
# mean values
mean_val = np.array([3, 2])

# the covariance matrices for the uncorrelated / correlated case
covMat_ucor = np.array([[25, 0], [0, 9]])
covMat_cor = np.array([[10, 5], [5, 5]])

# the x,y grid
x, y = np.mgrid[-4:10:0.2, -4:8:0.2]
pos = np.dstack((x,y))

# the pdfs
rv_ucor = stats.multivariate_normal(mean_val, covMat_ucor)
rv_cor = stats.multivariate_normal(mean_val, covMat_cor)

# evaluating the density functions
pdf_ucor = rv_ucor.pdf(pos)
pdf_cor = rv_cor.pdf(pos)

levels_ucor = np.linspace(np.min(pdf_ucor), np.max(pdf_ucor), 10)
levels_cor = np.linspace(np.min(pdf_cor), np.max(pdf_cor), 10)
```

```python
fig1, ax1 = plt.subplots(nrows=1, ncols=2, figsize=(10,8))
fig1.set_tight_layout(True)

ax1[0].set_aspect('equal', 'box')
#ax1[0].imshow( (x, y, pdf_ucor))
ax1[0].contourf(x, y, pdf_ucor, cmap='Greys', levels=levels_ucor)
ax1[0].axvline(mean_val[0], color='r', linestyle=':')
ax1[0].axhline(mean_val[1], color='r', linestyle=':')
ax1[0].set_xlabel('x')
ax1[0].set_ylabel('y')
ax1[0].set_title('')
#ax1[0].legend()
ax1[0].grid(True)
#ax1[0].set_xlim(np.min(x), np.max(x))
#ax1[0].set_ylim(np.min(y), np.max(y))

ax1[1].set_aspect('equal', 'box')
ax1[1].contourf(x, y, pdf_cor, cmap='Greys', levels=levels_cor)
ax1[1].axvline(mean_val[0], color='r', linestyle=':')
ax1[1].axhline(mean_val[1], color='r', linestyle=':')
ax1[1].set_xlabel('x')
ax1[1].set_ylabel('y')
ax1[1].set_title('')
#ax1[1].legend()
ax1[1].grid(True) 
#ax1[1].set_xlim(np.min(x), np.max(x))
#ax1[1].set_ylim(np.min(y), np.max(y))
```

## The Covariance Matrix as a mixing process of uncorrelated random variables

The properties of the covariance matrix

1) real and symmetric

2) positive definite

allow to transform the covariance matrix into a product of three matrices:

$$
\mathbf{\Sigma} = \mathbf{U} \cdot \mathbf{\Lambda} \cdot \mathbf{U}^T
$$

| symbol | description |
|--------|-------------|
| $\mathbf{\Sigma}$ | covariance matrix |
| $\mathbf{U}$ | orhogonal matrix of eigenvectors as column vectors |
| $\mathbf{\Lambda}$ | diagonal matrix of eigenvalues |

Matrix $\mathbf{\Lambda} $ can be factored like this:

$$
\mathbf{\Lambda}  = \mathbf{\Lambda^{1/2}} \cdot \mathbf{\Lambda^{1/2}}^T
$$

$\mathbf{\Lambda^{1/2}}$ is again a diagonal matrix with the square roots of eigenvalues.

$$\begin{align}
\mathbf{\Sigma} &= \mathbf{U} \cdot \mathbf{\Lambda^{1/2}} \cdot \mathbf{\Lambda^{1/2}}^T \cdot \mathbf{U}^T \\
&= \left(\mathbf{U} \cdot \mathbf{\Lambda^{1/2}} \right) \cdot \left( \mathbf{U} \cdot \mathbf{\Lambda^{1/2}} \right)^T \\
&= \mathbf{B} \cdot \mathbf{B}^T
\end{align}
$$

with $\mathbf{B} = \mathbf{U} \cdot \mathbf{\Lambda^{1/2}}$

Since $\mathbf{U}$ has orthogonal column vectors each column vector of $\mathbf{B}$ is a scaled column vector of $\mathbf{U}$. So the column vectors of $\mathbf{B}$ are orthogonal too.

The inverse covariance matrix is computed as:

$$
\mathbf{\Sigma^{-1}} = \left( \mathbf{B} \cdot \mathbf{B}^T  \right)^{-1} = \left( \mathbf{B}^T\right)^{-1} \cdot \mathbf{B}^{-1} = \mathbf{B}^{-T} \cdot \mathbf{B}^{-1}
$$

A numerical examples shall demonstrate these results:

We are using the covariance matrix

$$
\mathbf{\Sigma} = \left[\begin{array}{cc}
10 & 5 \\
5 & 5
\end{array}\right]
$$

1) compute the eigenvector decomposition and print eigenvalues and matrix of eigenvectors (column vectors)

2) show that the eigenvector decomposition is again the covariance matrix

3) compute the matrix $\mathbf{B}$ and show that the product $\mathbf{B} \cdot \mathbf{B}^T$ is again the covariance matrix

---

```python
# numerical example

covMat_cor = np.array([[10, 5], [5, 5]])

# computing the eigenvector decomposition of the covariance matrix

eigenvalues, eigenvectors = linalg.eigh(covMat_cor)

print(f"eigenvalues:\n{eigenvalues}")
print(f"eigenvectors:\n{eigenvectors}")
```

```python
# to show that the eigenvector decomposition yields the covariance matrix covMat_cor
mat = eigenvectors @ np.diag(eigenvalues) @ eigenvectors.T
mat
```

```python
B_mat = eigenvectors @ np.diag(np.sqrt(eigenvalues))
B_mat
```

```python
# the product is again the covariance matrix
B_mat @ B_mat.T
```
