```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import math
```

```python
std_degree = 15.0
std_range = 0.02

# conversion to std_rad
std_radians = math.radians(std_degree)

N = 100000
angles_vec = stats.norm.rvs(loc=0.5*math.pi, scale=std_radians, size=N)
radius_vec = stats.norm.rvs(loc=1.0, scale=std_range, size=N)
```

```python
x = radius_vec * np.cos(angles_vec)
y = radius_vec * np.sin(angles_vec)
```

```python
fig1, ax1 = plt.subplots(nrows=1, ncols=2, figsize=(10,4))

fig1.tight_layout()

ax1[0].scatter(x, y, color='g', s=1, marker='.', label='xy')
ax1[0].set_xlabel('x [m]')
ax1[0].set_ylabel('y [m]')
ax1[0].axis('equal')
ax1[0].set_title('')
ax1[0].legend()
ax1[0].grid(True)

ax1[1].scatter(angles_vec, radius_vec, color='g', s=1, marker='.', label='xy')
ax1[1].set_xlabel('angle [rad]')
ax1[1].set_ylabel('radius [m]')
# ax1[1].axis('equal')
ax1[1].set_title('')
ax1[1].legend()
ax1[1].grid(True)
```

```python
# print some statistics

print(f"mean(x): {np.mean(x):.3e}")
print(f"mean(y): {np.mean(y):.3e}")
print(f"mean(angle): {math.degrees(np.mean(angles_vec)):.3e} degree")
print(f"mean(radius): {np.mean(radius_vec):.3e}")
```

The conversion from polar into cartesian coordinates impacts the statistics. 

From the polar data we would obtain a mean value of 1 for the y coordinate. This is slightly larger than what would be obtained from the statistics in cartesian coordinates.
