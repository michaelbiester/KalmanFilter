For the linear case the measurement equation of the Kalman filter is given by this equation:

$$
\mathbf{z_n} = \mathbf{H} \cdot \mathbf{x_n} + \mathbf{v_n}
$$

Is there a way to estimate the state variable $\mathbf{x_n}$ ? 

Let $\mathbf{\hat{x}}_n$ be the estimated state. 

Let us assume that the linear estimate is found from 

$$
\mathbf{\hat{x}}_n = \mathbf{A} \cdot \mathbf{z_n} + \mathbf{b}
$$

How should be choose $\mathbf{A}$ and $\mathbf{b}$ ?

We define the error $\mathbf{e}_n$ as :

$$\begin{align}
\mathbf{e}_n &= \mathbf{x}_n - \mathbf{\hat{x}}_n \\
&= \mathbf{x}_n - \mathbf{A} \cdot \mathbf{z}_n - \mathbf{b}
\end{align}
$$

and which to minimise the expectation

$$
E\left(\mathbf{e}_n^T \cdot \mathbf{e}_n  \right)
$$

$$\begin{align}
E\left(\mathbf{e}_n^T \cdot \mathbf{e}_n  \right) &=  \\
&= E\left( \left( \mathbf{x}_n - \mathbf{A} \cdot \mathbf{z}_n - \mathbf{b} \right)^T \cdot \left( \mathbf{x}_n - \mathbf{A} \cdot \mathbf{z}_n - \mathbf{b} \right) \right) \\
&= E\left(\mathbf{x}_n^T \cdot \mathbf{x}_n \right) - E\left(\mathbf{x}_n^T \cdot \mathbf{A} \cdot \mathbf{z}_n \right) - E\left(\mathbf{x}_n^T \cdot \mathbf{b} \right) \\
&- E\left(\mathbf{z}_n^T \cdot \mathbf{A}^T \cdot \mathbf{x}_n \right) + E\left(\mathbf{z}_n^T \cdot \mathbf{A}^T \cdot \mathbf{A} \cdot \mathbf{z}_n\right) + E\left(\mathbf{z}_n^T \cdot \mathbf{A}^T \cdot \mathbf{b} \right) \\
&-E\left(\mathbf{b}^T \cdot \mathbf{x}_n \right) + E\left(\mathbf{b}^T \cdot \mathbf{A} \cdot \mathbf{z}_n \right) + E\left(\mathbf{b}^T \cdot\mathbf{b} \right)
\end{align}
$$

Now we need to compute the following derivatives:

$$
\frac{\partial E\left(\mathbf{e}_n^T \cdot \mathbf{e}_n  \right)}{\partial \mathbf{b}} = E\left(\frac{\partial \mathbf{e}_n^T \cdot \mathbf{e}_n }{\partial \mathbf{b}}  \right)
$$

and

$$
\frac{\partial E\left(\mathbf{e}_n^T \cdot \mathbf{e}_n  \right)}{\partial \mathbf{A}} = E\left(\frac{\partial \mathbf{e}_n^T \cdot \mathbf{e}_n }{\partial \mathbf{A}}  \right)
$$

Thereby we assume that we may first compute the derivatives and then take the expectation.

$$
E\left(\frac{\partial \mathbf{e}_n^T \cdot \mathbf{e}_n }{\partial \mathbf{b}}  \right) =  - E\left(\frac{\partial \mathbf{x}_n^T \cdot \mathbf{b}}{\partial \mathbf{b}} \right) +  E\left(\frac{\partial \mathbf{z}_n^T \cdot \mathbf{A}^T \cdot \mathbf{b}}{\partial \mathbf{b}} \right) -E\left(\frac{\partial \mathbf{b}^T \cdot \mathbf{x}_n}{\partial \mathbf{b}} \right) + E\left(\frac{\partial \mathbf{b}^T \cdot \mathbf{A} \cdot \mathbf{z}_n}{\partial \mathbf{b}} \right) + E\left(\frac{\partial \mathbf{b}^T \cdot\mathbf{b}}{\partial \mathbf{b}} \right)
$$

$$\begin{align}
E\left(\frac{\partial \mathbf{x}_n^T \cdot \mathbf{b}}{\partial \mathbf{b}} \right) &= E\left(\mathbf{x}_n \right)\\
E\left(\frac{\partial \mathbf{z}_n^T \cdot \mathbf{A}^T \cdot \mathbf{b}}{\partial \mathbf{b}} \right) &= E\left( \mathbf{A} \cdot \mathbf{z}_n \right) = \mathbf{A} \cdot  E\left( \mathbf{z}_n \right) \\
E\left(\frac{\partial \mathbf{b}^T \cdot \mathbf{x}_n}{\partial \mathbf{b}} \right) &= E\left(\mathbf{x}_n \right) \\
E\left(\frac{\partial \mathbf{b}^T \cdot \mathbf{A} \cdot \mathbf{z}_n}{\partial \mathbf{b}} \right) &= E\left( \mathbf{A} \cdot \mathbf{z}_n \right) = \mathbf{A} \cdot  E\left( \mathbf{z}_n \right)\\
E\left(\frac{\partial \mathbf{b}^T \cdot \mathbf{b}}{\partial \mathbf{b}} \right) &= 2 \cdot \mathbf{b}
\end{align}
$$

Putting these derivatives together and setting $E\left(\frac{\partial \mathbf{x}_n^T \cdot \mathbf{b}}{\partial \mathbf{b}} \right)$ to $\mathbf{0}$ yields:

$$
\mathbf{0} = - 2 \cdot E\left(\mathbf{x}_n \right) + \mathbf{A} \cdot  E\left( \mathbf{z}_n \right) + 2 \cdot \mathbf{b}
$$

Solving for $\mathbf{b}$:

$$
\mathbf{b}= E\left(\mathbf{x}_n \right) - \mathbf{A} \cdot  E\left( \mathbf{z}_n \right)
$$

And now a similar procedure to compute the matrix derivative:

$$\begin{align}
E\left(\frac{\partial \mathbf{e}_n^T \cdot \mathbf{e}_n }{\partial \mathbf{A}}  \right) &= -E\left(\frac{\partial \mathbf{x}_n^T \cdot \mathbf{A} \cdot \mathbf{z}_n}{\partial \mathbf{A}} \right) - E\left(\frac{\partial \mathbf{z}_n^T \cdot \mathbf{A}^T \cdot \mathbf{x}_n}{\partial \mathbf{A}} \right) + E\left(\frac{\partial \mathbf{z}_n^T \cdot \mathbf{A}^T \cdot \mathbf{A} \cdot \mathbf{z}_n }{\partial \mathbf{A}}\right) + E\left(\frac{\partial \mathbf{z}_n^T \cdot \mathbf{A}^T \cdot \mathbf{b} }{\partial \mathbf{A}} \right) + E\left(\frac{\partial \mathbf{b}^T \cdot \mathbf{A} \cdot \mathbf{z}_n }{\partial \mathbf{A}}\right)
\end{align}
$$

$$\begin{align}
E\left(\frac{\partial \mathbf{x}_n^T \cdot \mathbf{A} \cdot \mathbf{z}_n}{\partial \mathbf{A}} \right) &= E\left(\mathbf{x}_n \cdot \mathbf{z}_n^T \right)\\
E\left(\frac{\partial \mathbf{z}_n^T \cdot \mathbf{A}^T \cdot \mathbf{x}_n}{\partial \mathbf{A}} \right) &= E\left(\mathbf{x}_n \cdot \mathbf{z}_n^T \right)\\
E\left(\frac{\partial \mathbf{z}_n^T \cdot \mathbf{A}^T \cdot \mathbf{A} \cdot \mathbf{z}_n }{\partial \mathbf{A}}\right) &= 2 \cdot \mathbf{A} \cdot E\left(\mathbf{z}_n \cdot \mathbf{z}_n^T \right) \\
E\left(\frac{\partial \mathbf{z}_n^T \cdot \mathbf{A}^T \cdot \mathbf{b} }{\partial \mathbf{A}} \right) &= E\left(\mathbf{b} \cdot \mathbf{z}_n^T \right)\\
E\left(\frac{\partial \mathbf{b}^T \cdot \mathbf{A} \cdot \mathbf{z}_n }{\partial \mathbf{A}}\right) &= E\left(\mathbf{b} \cdot \mathbf{z}_n^T \right)
\end{align}
$$

Putting these derivatives together and setting $E\left(\frac{\partial \mathbf{e}_n^T \cdot \mathbf{e}_n }{\partial \mathbf{A}}  \right) = \mathbf{0}$ yields:

$$\begin{align}
E\left(\mathbf{x}_n \cdot \mathbf{z}_n^T \right) &= \mathbf{A} \cdot E\left(\mathbf{z}_n \cdot \mathbf{z}_n^T \right) + E\left(\mathbf{b} \cdot \mathbf{z}_n^T \right) \\
&= \mathbf{A} \cdot E\left(\mathbf{z}_n \cdot \mathbf{z}_n^T \right) + \mathbf{b} \cdot  E\left(\mathbf{z}_n^T \right) \\
&= \mathbf{A} \cdot E\left(\mathbf{z}_n \cdot \mathbf{z}_n^T \right) + E\left(\mathbf{x}_n \right) \cdot  E\left(\mathbf{z}_n^T \right) - \mathbf{A} \cdot  E\left( \mathbf{z}_n \right) \cdot  E\left(\mathbf{z}_n^T \right)
\end{align}
$$

$$\begin{align}
E\left(\mathbf{x}_n \cdot \mathbf{z}_n^T \right) - E\left(\mathbf{x}_n \right) \cdot  E\left(\mathbf{z}_n^T \right) &= \mathbf{A} \cdot \left( E\left(\mathbf{z}_n \cdot \mathbf{z}_n^T \right) - E\left( \mathbf{z}_n \right) \cdot  E\left(\mathbf{z}_n^T \right) \right) \\
E\left( \left(\mathbf{x}_n - E\left(\mathbf{x}_n \right) \right) \cdot \left(\mathbf{z}_n - E\left(\mathbf{z}_n \right) \right)^T \right) &= \mathbf{A} \cdot E\left( \left(\mathbf{z}_n - E\left(\mathbf{z}_n \right) \right) \cdot \left(\mathbf{z}_n - E\left(\mathbf{z}_n \right) \right)^T \right)
\end{align}
$$

$$
\mathbf{P}_{x,z} = \mathbf{A} \cdot \mathbf{P}_{z,z}
$$

$$
\mathbf{A} = \mathbf{P}_{x,z} \cdot \mathbf{P}_{z,z}^{-1}
$$

$$\begin{align}
\mathbf{\hat{x}}_n &= \mathbf{A} \cdot \mathbf{z_n} + \mathbf{b} \\
&= \mathbf{A} \cdot \left(\mathbf{z_n} - E\left(\mathbf{z_n}\right) \right) \\
&= \mathbf{P}_{x,z} \cdot \mathbf{P}_{z,z}^{-1} \cdot \left(\mathbf{z_n} - E\left(\mathbf{z_n}\right) \right)
\end{align}
$$
